```python
import re, time, json
import pandas as pd
import requests
from bs4 import BeautifulSoup
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

URL = "https://iranhrs.org/%D8%A7%D8%B3%D8%A7%D9%85%DB%8C-%D9%88-%D9%85%D8%B4%D8%AE%D8%B5%D8%A7%D8%AA-%DB%B6%DB%B6%DB%B2-%D8%B4%D9%87%DB%8C%D8%AF-%D8%B1%D8%A7%D9%87-%D8%A2%D8%B2%D8%A7%D8%AF%DB%8C-%D8%AF%D8%B1-%D8%A7%D8%B9%D8%AA/"

OUT_CSV = "iranhrs_662_extracted.csv"
CACHE_JSON = "geocode_cache_iran.json"

PERSIAN_DIGITS = str.maketrans("۰۱۲۳۴۵۶۷۸۹", "0123456789")

def norm(s: str) -> str:
    s = (s or "").replace("\u200c", "")
    s = s.translate(PERSIAN_DIGITS)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def extract_article_text(html: str) -> str:
    soup = BeautifulSoup(html, "html.parser")
    # WordPress: usually <article> or .entry-content
    node = soup.select_one(".entry-content") or soup.select_one("article") or soup.body
    txt = node.get_text("\n", strip=True)
    return norm(txt)

def split_records(text: str):
    # matches: "123- ...." (hyphen variants)
    pat = re.compile(r"(?m)^\s*(\d{1,4})\s*[-–ـ]\s*(.+?)(?=^\s*\d{1,4}\s*[-–ـ]\s*|\Z)", re.DOTALL)
    for m in pat.finditer(text):
        no = int(m.group(1))
        body = norm(m.group(2))
        yield no, body

def parse_fields(body: str):
    # Most rows look like: "نام، شهر، ۳۵ ساله، ..."
    parts = [p.strip() for p in body.split("،") if p.strip()]
    name_fa = parts[0] if parts else ""

    age = None
    m = re.search(r"(\d{1,3})\s*ساله", body)
    if m:
        age = int(m.group(1))

    city_fa = ""
    m = re.search(r"(?:ساکن|اهل)\s+([^\s،\(\)]+)", body)
    if m:
        city_fa = m.group(1)
    elif len(parts) >= 2:
        city_fa = parts[1]

    return name_fa, age, city_fa

# gender inference (best-effort)
female_first_names = {
    "مهسا","نیکا","حدیث","سارینا","ندا","زهرا","فاطمه","مریم","لیلا","سحر","شبنم","پانیز","آزاده","مینو","الهه","ستاره","نازنین","پریسا","شیرین","حنانه"
}
def infer_gender(name_fa: str, body: str) -> str:
    if any(k in body for k in ["زن", "دختر", "خانم", "بانوی", "مادر"]):
        return "Female"
    first = norm(name_fa).split()[0] if name_fa else ""
    return "Female" if first in female_first_names else "Male"

# geocoding with cache + rate limit
def load_cache(path):
    try:
        with open(path, "r", encoding="utf-8") as f:
            return json.load(f)
    except FileNotFoundError:
        return {}

def save_cache(path, cache):
    with open(path, "w", encoding="utf-8") as f:
        json.dump(cache, f, ensure_ascii=False, indent=2)

cache = load_cache(CACHE_JSON)

geolocator = Nominatim(user_agent="iranhrs_662_extractor")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1.1)  # be polite

IRAN_CENTER = (32.4279, 53.6880)

def geocode_city(city_fa: str):
    if not city_fa:
        return IRAN_CENTER
    key = f"{city_fa}, Iran"
    if key in cache:
        return tuple(cache[key])
    loc = geocode(key)
    if loc:
        cache[key] = [loc.latitude, loc.longitude]
        return loc.latitude, loc.longitude
    cache[key] = [IRAN_CENTER[0], IRAN_CENTER[1]]
    return IRAN_CENTER

# ---- run ----
html = requests.get(URL, timeout=60).text

text = textmanual #extract_article_text(html)

rows = []
for no, body in split_records(text):
    name_fa, age, city_fa = parse_fields(body)
    if not name_fa:
        continue
    gender = infer_gender(name_fa, body)
    lat, lon = geocode_city(city_fa)
    rows.append({
        "Name in Farsi": name_fa,
        "Name in English": "",     # optional: add transliteration later
        "Year": 2022,
        "Age": age,
        "Gender": gender,
        "City": city_fa,
        "Province": "",            # optional: add city->province map later
        "Country": "Iran",
        "latitude": lat,
        "longitude": lon,
    })

df = pd.DataFrame(rows).drop_duplicates()
df.to_csv(OUT_CSV, index=False, encoding="utf-8-sig")
save_cache(CACHE_JSON, cache)

print("Saved:", OUT_CSV, "| rows:", len(df))
```
